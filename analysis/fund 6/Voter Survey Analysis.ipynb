```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# importing data
xls = pd.ExcelFile('Cardano Catalyst Incentivized Voter Survey (Responses).xlsx')
df = pd.read_excel(xls, 'Cleaned Data (with new wallets)')
```

```python
column = 'In which previous funds have you voted?'
l = list(df[column])
l = [i.split(', ') for i in l]
l2 = []
for i in l:
    for n in range(len(i)):
        l2.append(i[n])
l2 = pd.Series(l2)
s1 = l2.value_counts()
fig = plt.figure(figsize=(10, 6))
ax = s1.plot.bar()
plt.title(column, y=1.025, fontsize = 12)
plt.savefig('./charts/' + column[:-1] + '.jpeg', dpi=300, bbox_inches="tight")
```

```python
column = 'How many proposals did you vote positive for?'
hue_column = 'Other than voting have you actively participated in Cardano Project Catalyst?'
def plot_hue(column, hue_column):
    fig = plt.figure(figsize=(10, 6))
    ax = sns.histplot(x = column, data = df.astype(str), hue = hue_column, multiple="stack", shrink=0.8)
    sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
    plt.xticks(rotation=90)
    plt.title(column, y=1.025, fontsize = 12)
    plt.savefig('./charts/' + column.replace('?','').replace('.','').replace('\"','') + '.jpeg', dpi=300, bbox_inches="tight")
    plt.show()
```

```python
def plot_column_value_counts(column):
    fig = plt.figure()
    df[column].value_counts().plot.bar()
    fig.suptitle(column)
    plt.savefig('./charts/' + column.replace('?','').replace('.','').replace('\"','') + '.jpeg', dpi=300, bbox_inches="tight")
    plt.show()
```

```python
columns = ['Other than voting have you actively participated in Cardano Project Catalyst?',
           'How many proposals did you vote positive for?',
           'How many proposals did you vote negative for?',
           'How much time did you spend on each proposal when considering it for your vote?',
           'Do you have an Ideascale account?',
           'In relation to Fund 6 voting, did you read the proposals you voted on in Ideascale.',
           'How important is getting rewarded for voting to you?',
           'In your opinion, the amount of voting reward is...',
           'How much weight do you give to Community Advisor scores?',
           'How much voting power do you control? (If you wish to answer and stay anonymous you can choose not to provide an email address.)',
           'How many wallets do you vote with?',
           'Do you vote on behalf of others?',
           'If there was a way to get "expert ballots" and "voter advice" that provide direction on your own voting choices, would you use them?',
           'In relation to voting would you prefer to spend more lore less time?',
           'How would you rate your general experience with voting in Catalyst Fund 6?',
           'If you have participated in any previous funds, did you perceive any improvement in the general process?',
           'How would you rate the importance of this or surveys like this?',
           'How would you rate our execution of this survey?',
           'How likely are you to participate in a voter survey if it was not incentivized?']

for i in columns:
    plot_column_value_counts(i)
```
