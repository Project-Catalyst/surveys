```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# importing data
xls = pd.ExcelFile('Portuguese - Private.xlsx')
df = pd.read_excel(xls, 'Respuestas de formulario 1')
```

```python
column = 'Em quais os fundos que vocÃª votou anteriormente?'
l = list(df[column])
l = [i.split(', ') for i in l]
l2 = []
for i in l:
    for n in range(len(i)):
        l2.append(i[n])
l2 = pd.Series(l2)
s1 = l2.value_counts()
fig = plt.figure(figsize=(10, 6))
ax = s1.plot.bar()
plt.title(column, y=1.025, fontsize = 12)
plt.savefig('./charts/' + column[:-1] + '.jpeg', dpi=300, bbox_inches="tight")
```

```python
column = 'How many proposals did you vote positive for?'
hue_column = 'Other than voting have you actively participated in Cardano Project Catalyst?'
def plot_hue(column, hue_column):
    fig = plt.figure(figsize=(10, 6))
    ax = sns.histplot(x = column, data = df.astype(str), hue = hue_column, multiple="stack", shrink=0.8)
    sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
    plt.xticks(rotation=90)
    plt.title(column, y=1.025, fontsize = 12)
    plt.savefig('./charts/' + column.replace('?','').replace('.','').replace('\"','').replace(':"','') + '.jpeg', dpi=300, bbox_inches="tight")
    plt.show()
```

```python
def plot_column_value_counts(column):
    fig = plt.figure()
    df[column].value_counts().plot.bar()
    fig.suptitle(column)
    plt.savefig('./charts/' + column.replace('?','').replace('.','').replace('\"','') + '.jpeg', dpi=300, bbox_inches="tight")
    plt.show()
```
